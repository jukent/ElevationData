```python
import pygmt
import geopandas as gpd
import rioxarray
```

```python
grid = pygmt.datasets.load_earth_relief(resolution="05m", region=[-130, -65, 24, 52])
grid.rio.write_crs("epsg:4326", inplace=True)
```

```python
path='tl_2023_us_state/tl_2023_us_state.shp'
df = gpd.read_file(path)

va_shape = df[df['NAME'] == 'Virginia']
va_shape = va_shape.to_crs("EPSG:4326")
```

```python
clipped = grid.rio.clip(va_shape.geometry.values, va_shape.crs, drop=False, invert=False)
```

```python
fig = pygmt.Figure()

region = [-85, -75, 36, 40]


# Set up the map
fig.coast(
    region=region,
    water="white",
)

# Plot the New York state boundary
fig.plot(data=va_shape, pen="0.5p", region=region, fill="#f8f8f8")


# Add contours (assuming `clipped` is the correct grid file)
fig.grdcontour(clipped, annotation=250, interval=50, region=region)

fig.text(text="\u2764",x=-77.0469, y=38.8048,font="15p,red")

# Display the figure
fig.show()
```
